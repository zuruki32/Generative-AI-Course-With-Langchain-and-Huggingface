```python
import re
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

# Download required NLTK data
nltk.download(['stopwords', 'wordnet', 'averaged_perceptron_tagger'])

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def get_wordnet_pos(treebank_tag):
    """Convert POS tags to WordNet format."""
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # Default to noun

corpus = []

for message in messages['message']:
    # Step 1: Clean text (remove punctuation, lowercase)
    text = re.sub('[^a-zA-Z]', ' ', message).lower()

    # Step 2: Tokenize and POS tag
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)  # e.g., [('running', 'VBG'), ('fast', 'JJ')]

    # Step 3: Lemmatize with POS
    lemmatized = []
    for word, tag in pos_tags:
        if word not in stop_words:
            wordnet_pos = get_wordnet_pos(tag)  # Convert to WordNet POS
            lemma = lemmatizer.lemmatize(word, pos=wordnet_pos)
            lemmatized.append(lemma)

    corpus.append(' '.join(lemmatized))  # Join back to a string
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf= TfidfVectorizer(max_features=500)
X=  tfidf.fit_transform(corpus).toarray()
```
